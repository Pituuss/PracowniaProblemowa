```python
@tf.function
def inner_fun(img_values, alpha):
    return tf.cast(1/(1-2*img_values)*(img_values - alpha), tf.float32)

@tf.function
def adjust_color(img_values, alpha, channel):
    current_channel_layer = img_values[:,channel]
    
    zeros = current_channel_layer < alpha
    fun = tf.cast(tf.cast(alpha <= current_channel_layer,tf.int8) * tf.cast(current_channel_layer < 1 - alpha,tf.int8),tf.bool)
    ones = current_channel_layer >= 1 - current_channel_layer
    
    img_proc_1 = tf.where(fun, inner_fun(current_channel_layer, alpha), 0.0)
    img_proc_2 = tf.where(ones, 1.0, img_proc_1)
    img_proc = tf.stack([ img_proc_2 if i is channel else img_values[:,i] for i in range(3)],axis=1)
    return img_proc

@tf.function
def adjust_color_merge(img, alphas, alpha_factor, channel):
    return tf.foldl(
        lambda acc, alpha: acc + alpha[0] * adjust_color(img, alpha[1], channel),
        tf.stack([alphas[0],alpha_factor],axis=1),
        parallel_iterations=2048,
        initializer=tf.zeros([256,3])
    ) / alphas.shape[1]

@tf.function
def img_to_set(img):
    vset = tf.map_fn(lambda x: _make_values_set(x), tf.reshape(img,(3,-1)), parallel_iterations=img.shape[0])
    return tf.stack([vset[:,i] for i in range(256)])

@tf.function
def _img_to_set(img_one_channel):
    uniq, _ = tf.unique(tf.reshape(img_one_channel,[-1]))
    return tf.concat([tf.sort(uniq), tf.zeros(256)],axis=0)[:256]

@tf.function
def _convert(img, old, new):
    return tf.foldl(lambda a,x: a + tf.where(img == x[0],x[1],0),tf.stack([old, new],axis=1),initializer=tf.zeros(img.shape))

@tf.function
def convert(images,oryginal,changed):
    return tf.map_fn(
            lambda i: _convert(images[tf.cast(i,tf.int32)], 
                          oryginal[tf.cast(i,tf.int32)],
                          changed[tf.cast(i,tf.int32)]),
            tf.range(8,dtype=tf.float32),
            parallel_iterations=images.shape[0])
```

```python
class AdjustColor_(Layer):

    def __init__(self, channel, trainable=True, **kwargs):
        super(AdjustColor_, self).__init__(**kwargs)
        self.supports_masking = True
        self.alpha = tf.random.normal([2048])
        self.trainable = trainable
        self.channel = channel

    def build(self, input_shape):
               
        self.alpha_factor = tf.Variable(self.alpha,
                                      shape=(2048),
                                      dtype='float32',
                                      name='alpha_factor')
        if self.trainable:
            self._trainable_weights.append(self.alpha_factor)

        super(AdjustColor_, self).build(input_shape)

    def call(self, images, features, mask=None):
        return tf.map_fn(lambda img: adjust_color_merge(img, features, self.alpha_factor, self.channel),
                         images,
                         parallel_iterations=images.shape[0]
                        )
```

```python
class ImgToSet(Layer):
    
    def __init__(self, trainable=False, **kwargs):
        super(ImgToSet, self).__init__(**kwargs)
        self.supports_masking = True
        
    def call(self, images, mask=None):
        return tf.map_fn(lambda img: img_to_set(img),images,parallel_iterations=images.shape[0])
```

```python
# Generator
G_inputs = tk.Input(shape=(100,), name=f"G_inputs")

x = tk.layers.Dense(7 * 7 * 16)(G_inputs)
x = tf.nn.leaky_relu(x)
x = tk.layers.Flatten()(x)
x = tk.layers.Reshape((7, 7, 16))(x)

x = tk.layers.Conv2DTranspose(32, (3, 3), padding="same")(x)
x = tk.layers.BatchNormalization()(x)
x = tf.nn.leaky_relu(x)
x = tf.image.resize(x, (14, 14))

x = tk.layers.Conv2DTranspose(32, (3, 3), padding="same")(x)
x = tk.layers.BatchNormalization()(x)
x = tf.nn.leaky_relu(x)
x = tf.image.resize(x, (28, 28))

x = tk.layers.Conv2DTranspose(32, (3, 3), padding="same")(x)
x = tk.layers.BatchNormalization()(x)
x = tf.nn.leaky_relu(x)

x = tk.layers.Conv2DTranspose(1, (3, 3), padding="same")(x)
x = tf.nn.sigmoid(x)

G_model = tk.Model(inputs=G_inputs,
                   outputs=x,
                   name="G")
G_model.summary()

# Discriminator
D_inputs = tk.Input(shape=(28, 28, 1), name=f"D_inputs")

x = tk.layers.Conv2D(32, (3, 3), padding="same")(D_inputs)
x = tf.nn.leaky_relu(x)
x = tk.layers.MaxPooling2D((2, 2))(x)
x = tk.layers.Conv2D(32, (3, 3), padding="same")(x)
x = tf.nn.leaky_relu(x)
x = tk.layers.MaxPooling2D((2, 2))(x)
x = tk.layers.Conv2D(64, (3, 3), padding="same")(x)
x = tf.nn.leaky_relu(x)

x = tk.layers.Flatten()(x)

x = tk.layers.Dense(128)(x)
x = tf.nn.sigmoid(x)
x = tk.layers.Dense(64)(x)
x = tf.nn.sigmoid(x)
x = tk.layers.Dense(1)(x)
x = tf.nn.sigmoid(x)

D_model = tk.Model(inputs=D_inputs,
                   outputs=x,
                   name="D")

D_model.compile(optimizer=tk.optimizers.Adam(learning_rate=1e-5, beta_1=0.5, name="Adam_D"),
                loss="binary_crossentropy")
D_model.summary()

GAN = tk.Sequential()
GAN.add(G_model)
GAN.add(D_model)
GAN.compile(optimizer=tk.optimizers.Adam(learning_rate=1e-5, beta_1=0.5, name="Adam_GAN"),
            loss="binary_crossentropy")

tb = tk.callbacks.TensorBoard(log_dir="./tb_tf2.0", write_graph=True)

# dummy data
noise = np.random.rand(100, 100).astype(np.float32)
target = np.ones(shape=(100, 1), dtype=np.float32)

GAN.fit(x=noise,
        y=target,
        callbacks=[tb])
```
